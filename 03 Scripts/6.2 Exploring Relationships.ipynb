# 03. Picking out variables for exploratory visual analysis

```python
# Create a subset with only quantitative columns

df_players_sub = df_players_cleaned[['age', 'height_cm', 'weight_kg', 'overall', 'potential', 'value_eur','wage_eur']]
```

# 04. Exploring relationships

# Correlations:

```python
# Create a correlation matrix using pandas

df_players_sub.corr()
```

```python
# Create a correlation heatmap using matplotlib

plt.matshow(df_players_sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_players_sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_players_sub.shape[1]), df_players_sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df_players_sub.shape[1]), df_players_sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

# Create a correlation heatmap using seaborn:

```python
# Create a subplot with matplotlib. What it does is create an empty output, but this is okay—it's just to show the subplots.

f,ax = plt.subplots(figsize=(10,10))
```

```python
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(df_players_sub.corr(), annot = True, ax = ax) 

# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

```python
corr.figure
```

weight_kg and height_cm have a strong relationship which definetely makes sense

value_eur and wage_eur have also a strong relationship. In general you can say: The higher the value of a player the higher the wage.

We also see a strong relationship between overall (FIFA 21 overall score) and value_eur. Same here: The higher overall score the higher the value in EUR.

We also see a strong relationship between overall (FIFA 21 overall score) and wage_eur. Same here: The higher overall score the higher the wage in EUR.

There is also a moderate relationship between overall and age.

# Scatterplots:

```python
# Create a scatterplot for the "value_eur" and "wage_eur" columns in seaborn

sns.lmplot(x = 'value_eur', y = 'wage_eur', data = df_players_cleaned)
```

There is a strong correlation between value and wage. In general you can say that the better the player the higher their monetary value as well as their wage.

```python
# Create a scatterplot for the "weight_kg" and "height_cm" columns in seaborn

sns.lmplot(x = 'weight_kg', y = 'height_cm', data = df_players_cleaned)
```

There is a strong correlation between weight and height due to the fact that taller people tend to be heavier.

# Pair Plots:

```python
# Keep only the variables you want to use in the pair plot

df_players_sub_2 = df_players_sub[['overall', 'value_eur', 'wage_eur', 'age']]
```

# Categorical Plots:

Creating a categorical variable that splits the "overall" column into categories

```python
# Use a histogram to visualize the distribution of the variables.

sns.distplot(df_players_cleaned['overall'], bins = 20)
```

```python
df_players_cleaned.loc[df_players_cleaned['overall'] < 60, 'fifa_score_category'] = 'Low score'
```

```python
df_players_cleaned.loc[(df_players_cleaned['overall'] >= 60) & (df_players_cleaned['overall'] < 75), 'fifa_score_category'] = 'Middle score'
```

```python
df_players_cleaned.loc[df_players_cleaned['overall'] >= 75, 'fifa_score_category'] = 'High score'
```

```python
df_players_cleaned['fifa_score_category'].value_counts(dropna = False)
```

```python
df_players_cleaned.columns
```

```python
# Create a categorical plot in seaborn using the fifa score categories created above

sns.set(style="ticks")
g = sns.catplot(x="value_eur", y="overall", hue="fifa_score_category", data=df_players_cleaned)
```

Here you can see that the better the overall score (player) the higher the monetary value of the player.
