# 03. Data cleaning

```python
# rename column "overall" to "overall_fifa_score"
df_players_cleaned.rename(columns = {'overall' : 'overall_fifa_score'}, inplace = True)
```

# 04. Changing a Variable‚Äôs Data Type

```python
# changing overall_fifa_score data type to integer
df_players_cleaned['overall_fifa_score'] = df_players_cleaned['overall_fifa_score'].astype('int')
```

```python
df_players_cleaned['overall_fifa_score'].dtype
```

# 05. Map setup

```python
# Create a data frame with just the nationalities and the overall fifa scores we want plotted

data_to_plot = df_players_cleaned[['nationality','overall_fifa_score']]
data_to_plot.head()
```

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
map.choropleth(geo_data = country_geo, data = data_to_plot,
             columns = ['nationality', 'overall_fifa_score'],
             key_on = 'Feature.properties.name', 
            # this part is very important - check your json file to see where the KEY is located
             fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
             legend_name = "overall_fifa_score")
```
